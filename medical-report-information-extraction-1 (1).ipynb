```python
import pandas as pd
import re

# âœ… Step 1: Load and Clean Dataset
df = pd.read_csv('/kaggle/input/medicaltranscriptions/mtsamples.csv')

# Drop rows without transcription
df = df.dropna(subset=['transcription']).reset_index(drop=True)

print(f"âœ… Total valid records: {len(df)}")

# âœ… Step 2: Preprocessing Function
def preprocess_text(text):
    # Lowercase
    text = text.lower()
    # Remove special characters (keep letters, digits, spaces, and some punctuation)
    text = re.sub(r"[^a-zA-Z0-9\s.,;:]", "", text)
    # Replace multiple spaces with one
    text = re.sub(r"\s+", " ", text).strip()
    return text

# âœ… Step 3: Process All Rows
all_clean_notes = []

for idx, row in df.iterrows():
    diagnosis = row['sample_name'] if pd.notnull(row['sample_name']) else "Unknown"
    raw_note = row['transcription']
    cleaned_note = preprocess_text(raw_note)
    
    all_clean_notes.append({
        "id": idx,
        "diagnosis": diagnosis.strip(),
        "note": cleaned_note
    })

# âœ… Step 4: Convert to DataFrame or Save as Needed
clean_df = pd.DataFrame(all_clean_notes)
print(clean_df.head(3))  # Show first 3 cleaned notes

# Optional: Save for later use
clean_df.to_csv("/kaggle/working/cleaned_transcriptions.csv", index=False)
print("\nâœ… Cleaned dataset saved as cleaned_transcriptions.csv")
```

```python
import spacy
import pandas as pd

# âœ… Load cleaned transcriptions (from previous step)
clean_df = pd.read_csv("/kaggle/working/cleaned_transcriptions.csv")

# âœ… Load biomedical spaCy model
nlp = spacy.load("en_ner_bionlp13cg_md")

# âœ… Create storage for results
entity_records = []

# âœ… Process each note
for idx, row in clean_df.iterrows():
    note_id = row['id']
    diagnosis = row['diagnosis']
    note = row['note']

    doc = nlp(note)

    entities = [(ent.text, ent.label_) for ent in doc.ents]

    entity_records.append({
        "id": note_id,
        "diagnosis": diagnosis,
        "entities": entities
    })

    if idx < 3:  # ğŸ‘€ Preview first few
        print(f"\nğŸ“ Note ID: {note_id}, Diagnosis: {diagnosis}")
        for ent in entities:
            print(f"  - {ent[0]} ({ent[1]})")

print("\nâœ… NER extraction complete for all notes.")

# âœ… Optional: Save all extracted entities
import json
with open("/kaggle/working/biomedical_entities.json", "w") as f:
    json.dump(entity_records, f, indent=2)

print("ğŸ§  Saved biomedical entities to biomedical_entities.json")
```

```python
import pandas as pd
import spacy
import json
import re
from tqdm import tqdm

# âœ… Load dataset
df = pd.read_csv('/kaggle/input/medicaltranscriptions/mtsamples.csv')
df = df.dropna(subset=['transcription']).reset_index(drop=True)

# âœ… Load biomedical NER model
nlp = spacy.load("en_ner_bionlp13cg_md")

# âœ… Define label-to-category mapping
label_mapping = {
    "DISEASE": "Diagnosis",
    "PROBLEM": "Symptoms",     # SpaCy may not return this in bionlp13cg
    "SYMPTOM": "Symptoms",     # Rare, but kept for compatibility
    "SIMPLE_CHEMICAL": "Treatments",
    "TREATMENT": "Treatments",
    "CHEMICAL": "Treatments",
    "DRUG": "Treatments",
    "GENE_OR_GENE_PRODUCT": "Treatments",
    "MULTI_TISSUE_STRUCTURE": "Organs",
    "TISSUE": "Organs",
    "ORGANISM_SUBDIVISION": "Organs",
    "ORGANISM_SUBSTANCE": "Organs",
    "ANATOMICAL_SYSTEM": "Organs"
}

# âœ… Common symptom keywords for rule-based extraction
symptom_keywords = [
    "pain", "fever", "cough", "sneezing", "headache", "fatigue",
    "itching", "nausea", "vomiting", "dizziness", "chills",
    "shortness of breath", "weakness", "runny nose", "congestion"
]

# âœ… Initialize structured info list
structured_info_all = []

# âœ… Loop over each transcription
for idx, row in tqdm(df.iterrows(), total=len(df)):
    note_text = row['transcription']
    diagnosis = row['sample_name']
    
    # NLP-based entity extraction
    doc = nlp(note_text)
    entities = [(ent.text.strip(), ent.label_) for ent in doc.ents]

    # Initialize structure
    structured_info = {
        "id": idx,
        "Diagnosis": [diagnosis.strip()],
        "Symptoms": [],
        "Treatments": [],
        "Organs": []
    }

    # Map NLP entities to categories
    for ent_text, ent_label in entities:
        ent_label = ent_label.upper()
        category = label_mapping.get(ent_label)
        if category:
            structured_info[category].append(ent_text.strip())

    # Rule-based symptom detection
    lower_note = note_text.lower()
    for symptom in symptom_keywords:
        if re.search(r"\b" + re.escape(symptom) + r"\b", lower_note):
            structured_info["Symptoms"].append(symptom)

    # Remove duplicates
    for key in structured_info:
        if isinstance(structured_info[key], list):
            structured_info[key] = list(set(structured_info[key]))

    structured_info_all.append(structured_info)

    # Optional: Print a few samples
    if idx < 3:
        print(f"\nğŸ“‹ Structured Info for Note ID: {idx}")
        print(json.dumps(structured_info, indent=2))

# âœ… Save all structured info to JSON
with open("/kaggle/working/all_structured_info.json", "w") as f:
    json.dump(structured_info_all, f, indent=2)

print(f"\nâœ… Done! Extracted structured info for {len(structured_info_all)} notes and saved to all_structured_info.json")
```

```python
import pandas as pd
import re
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize, sent_tokenize

# Example transcription
sample_note = notes_df['transcription'].iloc[0]  # from medical transcription dataset

# Clean basic symbols
sample_note = re.sub(r'\n+', ' ', sample_note)
sample_note = re.sub(r'[^a-zA-Z0-9.,;:()\- ]', '', sample_note)
```

```python
DISEASE_KEYWORDS = ["asthma", "allergy", "infection", "cancer", "diabetes", "rhinitis"]
TREATMENT_KEYWORDS = ["claritin", "zyrtec", "nasonex", "medication", "tablet", "inhaler"]
ANATOMY_KEYWORDS = ["lungs", "heart", "brain", "liver", "nostril", "blood", "mucosa"]
SYMPTOM_KEYWORDS = ["cough", "sneeze", "pain", "breathless", "headache", "itchy"]

structured_info = {
    "Diagnosis": [],
    "Symptoms": [],
    "Treatments": [],
    "Organs": []
}

tokens = word_tokenize(sample_note.lower())

for token in tokens:
    if token in DISEASE_KEYWORDS:
        structured_info["Diagnosis"].append(token)
    if token in TREATMENT_KEYWORDS:
        structured_info["Treatments"].append(token)
    if token in ANATOMY_KEYWORDS:
        structured_info["Organs"].append(token)
    if token in SYMPTOM_KEYWORDS:
        structured_info["Symptoms"].append(token)

# Remove duplicates
for key in structured_info:
    structured_info[key] = list(set(structured_info[key]))

print(structured_info)
```
